```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/storage/scratch/groshchupkin/Tom_dataset/annotations.csv')
# print(data)
```

```python
unique_pathways = data['SUPER_PATHWAY'].unique()
sorted_unique_values = sorted(unique_pathways)
# print(sorted_unique_values)
prior_mask = pd.DataFrame(0, index=data['ID'], columns=sorted_unique_values)
# print(prior_mask)
```

```python
for row in data.iterrows():
    prior_mask.loc[row[1]['ID'], row[1]['SUPER_PATHWAY']] = 1
prior_mask_array = prior_mask.to_numpy()
```

```python
import scipy.sparse

sparse_matrix = scipy.sparse.csr_matrix(prior_mask_array)
scipy.sparse.save_npz('/storage/scratch/groshchupkin/Tom_dataset/prior_mask.npz', sparse_matrix)
```
